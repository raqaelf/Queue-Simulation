```python
def masuk(x,a,maks):
    if len(a) < maks:
        a.append(x)
        return True
    else:
        return False
```

```python
def keluar(x,a):
    if len(a)==0:
        return False
    z = a.pop(x)
    return z
```

```python
from random import randint
import time

blokA = []
blokB = []

antrian = 1
antrian_end = 30
end_time = 100
maks = 5
for i in range(1,end_time):
    if antrian <= antrian_end:
        if masuk(antrian,blokA,maks):
            print('antrian ',antrian,' masuk ke blok A')
            antrian += 1
        elif masuk(antrian,blokB,maks):
            print('antrian ',antrian,' masuk ke blok B')
            antrian += 1
        else:
            print('antrian sedang penuh...')
        print('A = ', blokA)
        print('B = ', blokB)
        print('----------------')
    
    if i % randint(1,3) == 0:
        if len(blokA) != 0 and len(blokB) != 0:
            if blokA[0] < blokB[0]:
                print('antrian ',keluar(0,blokA),' Telah keluar dari antrian')
            else:
                print('antrian ',keluar(0,blokB),' Telah keluar dari antrian')
        elif len(blokA) != 0 :
            print('antrian ',keluar(0,blokA),' Telah keluar dari antrian')
        elif len(blokB) != 0 :
            print('antrian ',keluar(0,blokB),' Telah keluar dari antrian')
        else:
            print('Antrian Kosong...')
            break
        print('A = ', blokA)
        print('B = ', blokB)
        print('----------------')
            
            
    time.sleep(1)
```
